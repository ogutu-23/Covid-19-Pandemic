## Correlation Coefficients Interpretation

### Total Cases and Total Deaths (0.618891):
There is a moderate positive correlation between total cases and total deaths. This suggests that as the number of total cases increases, the number of total deaths tends to increase as well. This is expected in a pandemic context, where more cases can lead to more deaths.

### Total Cases and Lockdown Restrictions (0.068698):
The correlation is very weak and positive. This indicates that there is almost no relationship between the total number of cases and the severity of lockdown restrictions. This could suggest that lockdown measures may not have been effective in reducing the number of cases, or that other factors are at play.

### Total Cases and Vaccination Rate (0.015601):
The correlation is very close to zero, indicating no significant relationship between total cases and vaccination rates in this dataset. This could imply that the vaccination rate did not have a noticeable impact on the total number of cases during the period analyzed.

### Total Cases and Hospital Capacity (0.016350):
Again, this correlation is very weak, suggesting that there is no significant relationship between total cases and hospital capacity. This could indicate that hospital capacity was not a limiting factor in the number of cases reported.

### Total Deaths and Lockdown Restrictions (-0.145168):
There is a weak negative correlation between total deaths and lockdown restrictions. This suggests that as lockdown restrictions become more severe, the total number of deaths may slightly decrease. However, the correlation is weak, so this relationship is not strong.

### Total Deaths and Vaccination Rate (0.012013):
This correlation is very close to zero, indicating no significant relationship between total deaths and vaccination rates in this dataset.

### Total Deaths and Hospital Capacity (0.006576):
This correlation is also very weak, suggesting no significant relationship between total deaths and hospital capacity.

### Lockdown Restrictions and Vaccination Rate (-0.035419):
There is a very weak negative correlation, indicating that there is almost no relationship between lockdown restrictions and vaccination rates.

### Lockdown Restrictions and Hospital Capacity (-0.029702):
This correlation is also very weak, suggesting no significant relationship between lockdown restrictions and hospital capacity.

### Vaccination Rate and Hospital Capacity (0.942477):
